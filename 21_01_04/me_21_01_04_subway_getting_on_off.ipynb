```python
import altair as alt
import pandas as pd

subway = pd.read_csv('서울시 지하철호선별 역별 승하차 인원 정보.csv', encoding='cp949')
subway = subway[:5000] 
brush = alt.selection_interval()

points = alt.Chart(subway).mark_point().encode(
    x='승차총승객수:Q',
    y='하차총승객수:Q',
    color=alt.condition(brush, '호선명:N', alt.value('lightgray'))
).add_selection(
    brush
)

bars = alt.Chart(subway).mark_bar().encode(
    y='호선명:N',
    color='호선명:N',
    x='count(호선명):Q'
).transform_filter(
    brush
)

points & bars
```
