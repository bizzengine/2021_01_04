```python
import altair as alt
import pandas as pd

subway = pd.read_csv('     .csv', encoding='cp949')
subway = subway[:5000] 
brush = alt.selection_interval()

points = alt.Chart(subway).mark_point().encode(
    x=':Q',
    y=':Q',
    color=alt.condition(brush, ':N', alt.value('lightgray'))
).add_selection(
    brush
)

bars = alt.Chart(subway).mark_bar().encode(
    y=':N',
    color=':N',
    x='count():Q'
).transform_filter(
    brush
)

points & bars
```
