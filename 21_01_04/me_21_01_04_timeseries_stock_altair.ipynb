```python
## pip install pandas-datareader
## pip install altair

import numpy as np
import pandas as pd
from pandas_datareader import data 
import altair as alt
```

```python
start = '2019-1-1'
end = '2020-12-31'
source = 'yahoo'

ageagle = data.DataReader("UAVS", start=start ,end=end, data_source=source).reset_index()
sifco = data.DataReader("SIF", start=start ,end=end, data_source=source).reset_index()
bridgeline = data.DataReader("BLIN", start=start ,end=end, data_source=source).reset_index()
```

```python
# return DataFrame

ageagle.head()
```

```python
# add stock name

ageagle['Symbol'] = 'UAVS'
sifco['Symbol'] = 'SIF'
bridgeline['Symbol'] = 'BLIN'
```

```python
# comb

stocks = pd.concat(
    [ageagle[['Date','Close','Volume','Symbol']],
     sifco[['Date','Close','Volume','Symbol']],
     bridgeline[['Date','Close','Volume','Symbol']]],
     axis=0)
```

```python
stocks.head()
```

```python
# basic line of UAVS

(alt.
  Chart(stocks[stocks.Symbol == 'UAVS']).
  mark_line().
  encode(x='Date', y='Close'))
```

```python
# all 

(alt.
  Chart(stocks).
  mark_line().
  encode(x='Date', y='Close', color='Symbol').
  properties(height=300, width=500))
```

```python
# resampling

blin_resampled = stocks[stocks.Symbol == 'BLIN'].resample('7D', on='Date').mean().reset_index()
```

```python
(alt.
  Chart(blin_resampled).
  mark_line().
  encode(x='Date', y='Close').
  properties(height=300, width=500))
```

```python
# multiple plots

price = (alt.
 Chart(stocks[stocks.Symbol == 'BLIN']).
 mark_line().
 encode(x='Date', y='Close'))
volume = (alt.
 Chart(stocks[stocks.Symbol == 'BLIN']).
 mark_line().
 encode(x='Date', y='Volume'))
price | volume
```

```python
# add average line 

line = (alt.
         Chart(stocks).
         mark_line().
         encode(x='Date', y='Close', color='Symbol').
         properties(height=400, width=500))
avg = (alt.
         Chart(stocks).
         mark_rule().
         encode(y='average(Close)', color='Symbol',   
                size=alt.value(2)))
line + avg
```
