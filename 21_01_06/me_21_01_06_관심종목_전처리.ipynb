```python
import os
import pandas as pd
pd.options.display.max_rows = 5000 
```

```python
os.listdir()
```

```python
os.listdir()

df = pd.DataFrame(columns=['종목코드', '종목명', '현재가', '거래소', '업종'])


for csv_file in os.listdir():
    if 'csv' in csv_file:
        if 'na' in csv_file: 
            df2 = pd.read_csv(csv_file, engine='python')
            df2 = df2[['종목코드', '종목명', '현재가']]
            df2['거래소'] = "나스닥" 
            df2['업종'] = csv_file[3:].split('.')[0]
            
        elif 'ny' in csv_file:
            df2 = pd.read_csv(csv_file, engine='python')
            df2 = df2[['종목코드', '종목명', '현재가']]
            df2['거래소'] = "뉴욕" 
            df2['업종'] = csv_file[3:].split('.')[0]
            
        else: 
            df2 = pd.read_csv(csv_file, engine='python')
            df2 = df2[['종목코드', '종목명', '현재가']]
            df2['거래소'] = "아맥스"
            df2['업종'] = csv_file[3:].split('.')[0]
    
    df = pd.concat([df, df2], ignore_index=True)
```

```python
df.to_csv('final.csv')
```
